## Visualization Tool: Matplotlib

Just like ggplot2 from R, there is a library in python that allows line plots, bar charts, pie charts, histogram, etc.

```python
#importing matplotlib for line plotting
import matplotlib.pyplot as plt
year = [1950, 1970, 1990, 2010]
pop = [2.529, 3.692, 5.263, 6.972]

plt.plot(year, pop) #plt.scatter is scatter plot
#plt.xscale('log') would be used for scatter plots with weird shapes
plt.show()
```

```python
#pwd is to look at current working directories of this jupyter notebook. So we can know where we can import from
#importing csv file into Jupyter Notebook
import pandas as pd

#Day 2.csv is in my current working directory, so I could technically just use "Day 2.csv"
#but this is to demonstrate that you can load from wherever you saved the file from
data = pd.read_csv("/Users/sunny/Desktop/DataLyfe/100-days-of-code/Code Log/Day 2.csv")
data
```

```python
#loc and iloc is froma pandas that allows us subset a dataframe, but that's for later to really review
Male_data = data.loc[(data.Gender == "Male")]
Male_height = Male_data["Height"]
```

```python
# Histogram of male height data from kaggle, higher the bin, narrow each block is.
plt.hist(Male_height, bins=50)

# Show and clear plot
plt.show()
plt.clf() #clearing plot for next graph

# Histogram of life_exp1950, 15 bins
plt.hist(Male_height, bins=15)

# Show and clear plot again to see how setting bin to 15 would look like. 
# Note that they all look like a normal curve, which is seems correct for distribution of male heights.
# I'm 75 inches btw
plt.show()

```

```python
#subsetting for height and weight from the data. looking quite similar to R.
Male_HnW = Male_data[["Height", "Weight"]]
#scatterplot of male height and weight
plt.scatter(x = Male_HnW.Height, y = Male_HnW.Weight, alpha = 0.4)
plt.xscale('log') #if the data more scattered, we could use log
plt.xlabel('Height') #labeling x
plt.ylabel('Weight') #labeling y
plt.title('Height and Weight of US Male') #adding title
plt.xticks([61,66,71,76], ["61 in", "66 in", "71 in", "76 in"]) #more features
# Add grid() call
plt.grid(True) #more features

# Show the plot
plt.show()
```
